### Step 6. Is there more male or female names in the dataset?

```python
gender_counts = baby_names['Gender'].value_counts()
print(gender_counts)
```

### Step 7. Group the dataset by name and assign to names

```python
baby_names.head()
```

```python
names = baby_names.groupby('Name')
print(names)
```

### Step 8. How many different names exist in the dataset?

```python
num_names = len(names.groups)
print(num_names)
```

### Step 9. What is the name with most occurrences?

```python
most_common_name = names.size().idxmax()
print(most_common_name)
```

### Step 10. How many different names have the least occurrences?

```python
min_occurence = names.size().min()
print(min_occurence)
names_with_min_occurence = (names.size() == min_occurence).sum()
print(names_with_min_occurence)
```

### Step 11. What is the median name occurrence?

```python
name_occurences = names.size()
median_occurrence = name_occurences.median()
print(median_occurrence)
```

### Step 12. What is the standard deviation of names?

```python
name_occurences = names.size()
#print(name_occurences)
std = name_occurences.std()
print(std)
```
