*  Rate of deaths/confirmed/Recovery cases
* Ranking countries according to the effects of covid based on death, recovery, confirmed
* Trend of covid as monthly/yearly: reducing or increasing
*  Mapping countires/provinces that have been affected by covid





```python
agg_data = covid.groupby('Country/Region').agg({
    'Confirmed':'sum',
    'Deaths':'sum',
    'Recovered':'sum'
}).reset_index()
```

```python
agg_data.head()
```

```python
max_confirmed = agg_data['Confirmed'].max()
print(max_confirmed)
max_deaths = agg_data['Deaths'].max()
print(max_deaths)
max_recovered = agg_data['Recovered'].max()
print(max_recovered)
```

MAPPING OF COVID INFECTION WORLDWIDE

```python
import geopandas as gdp
```

```python
world = gdp.read_file(gdp.datasets.get_path('naturalearth_lowres'))
world.head()
```

```python
world = world.merge(agg_data, how='left', left_on='name', right_on='Country/Region')
world.head()
```

```python
fig, ax=plt.subplots(1, 1, figsize=(15, 10))
world.boundary.plot(ax=ax, linewidth=1)
world.plot(column='Confirmed', ax=ax, legend=True, legend_kwds={'label':'Confirmed Covid-19 Cases by Country', 'orientation':'horizontal'}, cmap='magma', missing_kwds={'color':'lightgrey'})
plt.title('Global Spread of COVID-19')
plt.show()
```
