### Step 2. Create the DataFrame with the following values:

### Step 3. Assign it to a variable called regiment.

### Step 6. What is the mean of each company's preTestScore?

```python
pretest_score = Regiment.groupby('company')['preTestScore'].mean()
print(pretest_score)
```

### Step 7. Present the mean preTestScores grouped by regiment and company

```python
'''
also known as multi-level indexing, is a way of organizing data in pandas where you have multiple levels of indices for both rows and columns.

In simpler terms, think of hierarchical indexing as having rows and columns organized in a structured way, where each row and column can have multiple levels of labels. It's like having subcategories within categories.
'''
```

```python
regiment_company = Regiment.groupby(['regiment', 'company'])['preTestScore'].mean()
print(regiment_company)
```

### Step 8. Present the mean preTestScores grouped by regiment and company without heirarchical indexing

```python
group = Regiment.groupby(['regiment', 'company'])['preTestScore'].mean().reset_index()
print(group)
```

### Step 9. Group the entire dataframe by regiment and company

```python
'''
When you iterate over a GroupBy object in pandas using a for loop, each iteration provides two values: the key (which represents the group name) and the group itself (which represents the subset of data associated with that group). This is why you can use for regiment_name, group in grouped_data: to iterate over the groups in a DataFrame grouped by a specific column.
'''
```

```python
group_data = Regiment.groupby(['regiment', 'company'])

#This line iterates over each group in the group_data object. For each group, it unpacks the group's index into two variables: regiment_name and company. It also creates a variable called group, which contains the data for that specific group.
for (regiment_name, company), group in group_data:
  print(regiment_name, company)
  print(group)
```

### Step 11. Iterate over a group and print the name and the whole data from the regiment

```python
grouped_data = Regiment.groupby('regiment')

#Iterate over each group and print the nameof the whole data
for regiment_name, group in grouped_data:
  print(regiment_name)
  print(group)
```
