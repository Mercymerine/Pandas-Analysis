### Step 5. Create a lambda function that will capitalize strings.

```python
# Example of a lambda function in comparison to normal function
'''
A lambda function is a small, anonymous function: meaning it doesn't have a name like a regular function does
Often used when you need a simple function for a short period of time and you don't want to use the keyword def
They can take any number of argumrnts, but they can only have one expression.
The syntax for a lambda function is lambda arguments: expression.
'''
#Function
def add(x, y):
  return x + y

#Equivalent lambda function
add_lambda = lambda x, y: x + y
result = add_lambda(3, 5)
print(result)
```

```python
capitalize_string = lambda x: x.capitalize() if isinstance(x, str) else x
```
